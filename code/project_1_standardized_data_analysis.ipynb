## Background

The SAT and ACT are standardized tests that many colleges and universities in the United States require for their admissions process. This score is used along with other materials such as grade point average (GPA) and essay responses to determine whether or not a potential student will be accepted to the university.

The SAT has two sections of the test: Evidence-Based Reading and Writing and Math ([*source*](https://www.princetonreview.com/college/sat-sections)). The ACT has 4 sections: English, Mathematics, Reading, and Science, with an additional optional writing section ([*source*](https://www.act.org/content/act/en/products-and-services/the-act/scores/understanding-your-scores.html)). [SAT](https://collegereadiness.collegeboard.org/sat) and [ACT](https://www.act.org/content/act/en.html) have different score range.

Since the 1940's, an increasing number of colleges have been using scores from students' performances on tests like the SAT and the ACT as a measure for college readiness and aptitude ([*source*](https://www.minotdailynews.com/news/local-news/2017/04/a-brief-history-of-the-sat-and-act/)). These tests scores being used as an objective measure to determine college admittance may not accurately measures the students potential or ability and serve as an inequitable barrier to entry. Since 2010, there has been a declining trend in the number of enrollments in postsecondary institutions, and the enrollment overall is between 66% and 70%. [[source]](https://www.usnews.com/news/education-news/articles/2019-04-25/college-enrollment-stays-flat-at-69-continuing-a-decade-long-trend) It was found that more than 60% of the students who took General Educational Development (GED) exam, in which the exam are accepted by most colleges, intended to attend college [[source]](https://educationdata.org/high-school-graduates-who-go-to-college) Therefore it is important to facilitate the students with the correct techniques to prepare for the SAT and/or ACT tests by providing SAT and ACT immersive programs to help and support more students to achieve a better score so that they can enter their desired college.

1. Manually calculate mean:

    Write a function that takes in values and returns the mean of the values. Create a list of numbers that you test on your function to check to make sure your function works!
    
    *Note*: Do not use any mean methods built-in to any Python libraries to do this! This should be done without importing any additional libraries.

```python
# Code:
def mean_values(list_values):
    num = 0
    total_number = len(list_values) # total number in the list
    
    for values in list_values:
        num += values # summation of the values in the list
    mean_num = (num/total_number)
    return mean_num
```

2. Manually calculate standard deviation:

    The formula for standard deviation is below:

    $$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

    Where $x_i$ represents each value in the dataset, $\mu$ represents the mean of all values in the dataset and $n$ represents the number of values in the dataset.

    Write a function that takes in values and returns the standard deviation of the values using the formula above. Hint: use the function you wrote above to calculate the mean! Use the list of numbers you created above to test on your function.
    
    *Note*: Do not use any standard deviation methods built-in to any Python libraries to do this! This should be done without importing any additional libraries.

```python
# Code:
def stand_dev(listvalue):
    summation = 0
    mean_num = mean_values(listvalue)
    total_numb = len(listvalue)
    
    
    for value in listvalue:
        summation += (value - mean_num)**2 #summation of the squared difference of the value from the mean
        
    sdev = ((1/total_numb)*summation)**0.5 # divide the summation with total number and square root it
    return sdev
```

    Write a function that takes in a string that is a number and a percent symbol (ex. '50%', '30.5%', etc.) and converts this to a float that is the decimal approximation of the percent.
```python
# Code:
def percent_to_float(percent):
    percent = percent.replace("%", "") # replace symbol '%' with no spaces
    number = (float(percent))/100 # convert the str into float and return in decimal approximation of the percent
    return number
```

**Using the National ACT Composite Scores and National SAT total scores to indicate the national average score in the scatterplot for comparison with the average state scores**

For ACT tests, the National United States ACT Composite scores are 21.0 (in 2017), 20.8 (in 2018), 20.7 (in 2019) (2017 & 2018 [source](https://nces.ed.gov/fastfacts/display.asp?id=897)) (2019 [source](https://nces.ed.gov/programs/digest/d19/tables/dt19_226.60.asp?current=yes)) 

For SAT tests, the National United States SAT total scores are 1060 (in 2017), 1068 (in 2018), 1059 (in 2019) (2017, 2018 & 2019 [source](https://nces.ed.gov/programs/digest/d19/tables/dt19_226.30.asp))

```python
# Define function to plot a scatterplot
def plot_scatterplot (data_list, xlist, ylist, labelx, labely, title_name, fig_range = (8,5), x_coordinates = [0,0], y_coordinates = [0,0], x_point = 0, y_point = 0, word = None):
    plt.figure(figsize = fig_range)
    sns.scatterplot(data = data_list, x = xlist, y = ylist)
    sns.set_style('whitegrid') # place gridlines
    plt.xlabel(labelx, fontdict = {'fontsize' : 15})
    plt.ylabel(labely, fontdict = {'fontsize' : 15})
    plt.yticks(fontsize=12)
    plt.xticks(fontsize=12)
    plt.title((title_name + '\n'), weight = 'bold', fontdict = {'fontsize' : 18})
    plt.plot(x_coordinates, y_coordinates, color='black', linestyle='dashed', linewidth=1, markersize=20) # plot straight dashed-line
    plt.text(x_point,y_point, word) # place a text
    return plt.show()
```

In USA, ACT and SAT tests are used as a standardized tests for college admissions with the idea behind to measure the college readiness of the students. As a result, there is a non-symmetrical distribution of participation rate and test scores for each test across the states in USA.
